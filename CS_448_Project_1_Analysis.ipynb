```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def createPlot(experiment, df):
  mru = df[df["mode"]=="MRU"]
  fault = df[df["mode"]=="default"]
  fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18,6))
  fig.suptitle(experiment + " hits and misses")

  # Hits Plot
  ax1.plot(mru["n"],mru["hits"], label="Most Recently Used")
  ax1.plot(fault["n"],fault["hits"], label="Default")
  ax1.title.set_text(experiment + " Hits")
  ax1.legend()
  
  # Misses Plot
  ax2.plot(mru["n"],mru["misses"], label="Most Recently Used")
  ax2.plot(fault["n"],fault["misses"], label="Default")
  ax2.title.set_text(experiment + " Misses")
  ax2.legend()
  
  # Ratio
  ax3.plot(mru["n"],mru["hits"]/mru["misses"], label="Most Recently Used")
  ax3.plot(fault["n"],fault["hits"]/fault["misses"], label="Default")
  ax3.title.set_text(experiment + " Hits/Misses")
  ax3.set_ylim(top=1, bottom=0)
  ax3.legend()
  fig.show()
```

```python
df = pd.read_csv("createTable.csv")
createPlot("Create Table", df)
```

```python
df = pd.read_csv("joinTable.csv")
createPlot("Join Table", df)
```

```python
df = pd.read_csv("selectTable.csv")
createPlot("Select Table", df)
```

```python
df = pd.read_csv("updateTable.csv")
createPlot("Update Table", df)
```

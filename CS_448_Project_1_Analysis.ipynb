```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def createPlot(experiment, df):
  mru = df[df["mode"]=="MRU"]
  fault = df[df["mode"]=="default"]
  fig, (ax1, ax2, ax4) = plt.subplots(1, 3, figsize=(24,6))
  fig.suptitle(experiment + " hits and misses")

  # Hits Plot
  ax1.plot(mru["n"],mru["hits"], label="Most Recently Used", alpha=0.7)
  ax1.plot(fault["n"],fault["hits"], label="Default", alpha=0.7)
  ax1.title.set_text(experiment + " Hits")
  ax1.set(xlabel="Size of input (n)", ylabel="Number of Hits")
  ax1.legend()
  
  # Misses Plot
  ax2.plot(mru["n"],mru["misses"], label="Most Recently Used", alpha=0.7)
  ax2.plot(fault["n"],fault["misses"], label="Default", alpha=0.7)
  ax2.title.set_text(experiment + " Misses")
  ax2.set(xlabel="Size of input (n)", ylabel="Number of Misses")
  ax2.legend()
  
  # # Writes
  # ax3.plot(mru["n"],mru["writes"], label="Most Recently Used")
  # ax3.plot(fault["n"],fault["writes"], label="Default")
  # ax3.title.set_text(experiment + " Total Writes")
  # ax3.legend()

  # Ratio
  ax4.plot(mru["n"],mru["hits"]/mru["misses"], label="Most Recently Used", alpha=0.7)
  ax4.plot(fault["n"],fault["hits"]/fault["misses"],   label="Default", alpha=0.7)
  ax4.title.set_text(experiment + " Hits:Misses Ratio")
  ax4.set(xlabel="Size of input (n)", ylabel="Hits:Misses ratio")
  ax4.legend()
  fig.show()
```

```python
df = pd.read_csv("mru_createTable.csv")
createPlot("Create Table", df)
```

```python
df = pd.read_csv("mru_joinTable.csv")
createPlot("Join Table", df)
```
